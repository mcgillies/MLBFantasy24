```python
import pandas as pd
```

```python
stats = pd.read_csv("data/b_fpoints.csv", index_col = 0)
metrics = pd.read_csv("data/bmetrics24.csv")
metrics.head()
```

```python
stats.head()
```

```python
metrics = metrics.rename(columns = {'last_name, first_name':'Name'})
metrics = metrics.drop(columns = ['player_id'])
metrics['year'] = metrics['year'] + 1
metrics = metrics[metrics['year'] < 2024]
stats = stats[stats['year'] > 2015]

metrics.head()
```

```python
fpts = stats[['Name','Fpoints_G', 'year']]
fpts.head()
```

```python
both = fpts.merge(metrics, on = ['Name', 'year'])
both.head()
```

```python
both.to_csv("data/b_combined.csv")
```

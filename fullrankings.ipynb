```python
import pandas as pd
```

```python
pdata = pd.read_csv('data/2024pitcherpreds.csv', index_col = 0)
bdata = pd.read_csv('data/2024batterpreds.csv', index_col = 0)
```

```python
bothdata = pd.concat([pdata, bdata], ignore_index = True)
bothdata = bothdata.drop(columns = ['index'])
bothdata.head()
```

```python
rankdata = bothdata.sort_values(by = 'Fpoints', ascending = False)
rankdata['rank'] = bothdata['Fpoints'].rank(ascending = False)
rankdata.head(50)
```

```python
rankdata.to_csv('data/finalranks.csv')
```
